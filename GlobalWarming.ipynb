```
data = pd.read_csv(r"https://raw.githubusercontent.com/themightyNJ/MachineLearning/master/emission_data.csv")
data = data.drop(index = 227)
```

```
pycountry.countries.search_fuzzy(data.Country.iloc[0])[0].alpha_3
```

```
some_countries = ["England", "HerpaDerpaland", "Engla"]

def do_fuzzy_search(country):
    try:
        result = pycountry.countries.search_fuzzy(country)
        return result[0].alpha_3
    except:
        return np.nan

for country in some_countries:
    print(do_fuzzy_search(country))
```

```
data['country_code'] = data["Country"].apply(lambda country: do_fuzzy_search(country))
```

```
plot_df = data.dropna()
fig = px.choropleth(plot_df, locations="country_code",
                    color="1817",
                    hover_name="Country",
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```
